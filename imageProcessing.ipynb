RBG TO GRAYSCALE

```python
import cv2
```

```python
img2 = cv2.imread('C:\\Users\\hp\\Desktop\\GitHub\\ImageProcessing\\cristian-castillo-73pyV0JJOmE-unsplash-scaled.jpg')
```

```python
grayscale_img = cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img)
```

```python
print(grayscale_img.shape)
```

```python
cv2.imshow('Grayscale Image', grayscale_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('dog_bw_img.jpg',grayscale_img)
```

```python
img3 = mpimg.imread('C:\\Users\\hp\\Desktop\\GitHub\\ImageProcessing\\dog_bw_img.jpg')
```
